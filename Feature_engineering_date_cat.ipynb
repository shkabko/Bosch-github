```python
# Create 1st dataframe
raw_data = { 'company': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd','1st', '1st', '2nd', '2nd'], 
        'name': ['Miller', 'Maga', np.nan, np.nan, 'Cooze', np.nan, np.nan, 'Sone', 'Sloan', 'Piger', 'Riani', 'Ali'],
        'name2': ['Ali', 'Maga', 'Ali', np.nan, 'Cooze', np.nan, np.nan, 'Sone', 'Sloan', 'Piger', 'Riani', 'Ali'],     
        'preTestScore': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],
        'Response': [0, 0, 1, 1, 1, 0, 0, 0, 0, 1, 0, 0]}
visible = pd.DataFrame(raw_data, columns = ['company', 'name','name2', 'preTestScore', 'Response'])
visible
```

```python
## group by names and take a mean of the response
outcomes = visible.groupby('name')['Response'].mean().reset_index()
outcomes
```
